```python
# The other columns in continuous variable that can be capped under upper and lower bound values
for i in cont_cols:
    q1=df1[i].describe()["25%"]
    q3=df1[i].describe()["75%"]
    iqr=q3-q1
    u_bound=q3 + 1.5 * iqr
    l_bound=q1 - 1.5 * iqr
    df1[i]=np.where(df1[i]<l_bound,l_bound,df1[i])
    df1[i]=np.where(df1[i]>u_bound,u_bound,df1[i])
```
